```python
import numpy as np
import matplotlib.pyplot as plt
from itertools import product
from collections import defaultdict
```

```python
num_simulations = 100000
sums_count = {i: 0 for i in range(2, 13)}

for _ in range(num_simulations):
    cube1 = np.random.randint(1, 7)
    cube2 = np.random.randint(1, 7)
    dice_sum = cube1 + cube2
    sums_count[dice_sum] += 1

probabilities = {k: v / num_simulations for k, v in sums_count.items()}
```

```python
sums = list(probabilities.keys())
probs = list(probabilities.values())
plt.bar(sums, probs, color='blue')
plt.xlabel('Сума чисел на кубиках')
plt.ylabel('Ймовірність')
plt.title('Ймовірності сум чисел на кубиках (Метод Монте-Карло)')
plt.show()
```
